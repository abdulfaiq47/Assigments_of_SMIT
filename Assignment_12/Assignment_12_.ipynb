cleaning text

```python
import re
import string
```

remove eveythings in single funtion

```python
def clean_text(text):
  # --------------------------
  # need "import re & string "
  # ---------------------------
  text = text.lower()
  # remove HTML TAG
  patterns = re.compile('<.*?>')
  text =  patterns.sub(r'',text)
  # remove URL
  pattern = re.compile(r'https?://|S+|www\.\S+')
  text =  pattern.sub(r'', text)

  exclude = string.punctuation
  for char in exclude:
    text = text.replace(char, '')
  return text
```

```python
data = clean_text(m_data)
```

```python
data[: 200]
```

Now tokenizing the data, vectorize and add layers



```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
```

```python
w_index = tokenizer.word_index
```

```python
total_world = len(w_index) + 1
```

```python
total_world
```

```python
token_list = tokenizer.texts_to_sequences([data])[0]
```

```python
# token_list
```

```python
input_seq = []
for i in range(1, len(token_list)):
  pos = token_list[:i+1]
  input_seq.append(pos)
```

```python
max_sequence = max(len(i) for i in input_seq)
```

```python
input_sequences = pad_sequences(input_seq, maxlen=max_sequence, padding='pre')
```

```python
input_sequences[0]
```

```python
x, y = input_sequences[:,:-1], input_sequences[:,-1]
```

```python
import numpy as np

print(f"Shape of x: {x.shape}")
print(f"Shape of y: {y.shape}")
print(f"Max value in y: {np.max(y)}")
print(f"Total words (vocabulary size): {total_world}")
```

```python
x[0]
```

```python
y[0]
```

```python
total_world
```

```python
model= Sequential()
model.add(Embedding(total_world, 512, input_length=max_sequence-1))
model.add(LSTM(128))
model.add(Dense(total_world, activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x,y,epochs=4, validation_split=0.2)
```

```python
def predict(model, tokenizer, text, max_sequence_len):
  token_list = tokenizer.texts_to_sequences([text])[0]

  token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')

  prediction = model.predict(token_list)

  prediction_word_index = np.argmax(prediction)

  for word, index in tokenizer.word_index.items():
    if index == prediction_word_index:
      return word

  return None
```

```python
text = "hy how"
for i in range(20):
  next_word = predict(model, tokenizer, text, max_sequence)
  text += " " + next_word
  if next_word is None:
    break

print(text)
```
